Попередня обробка має включати

Напишіть код аби виконати це завдання. Перед цим рекомендую детально ознайомитись з тим, що робить обʼєкт `sklearn.feature_extraction.text.CountVectorizer` за замовченням.

```python
stemmer = SnowballStemmer(language='english')
# Stem the stop words to make them consistent with the text processing
english_stopwords = [stemmer.stem(stopword) for stopword in stopwords.words('english')]
```

```python
# Custom tokenizer function
def tokenize(text):
    # Lowercase the text
    text = text.lower()
    # Remove URLs, mentions, hashtags, and other common patterns
    text = re.sub(r"http\S+|www\S+|@\w+|#\w+", "", text)
    # Tokenize and stem the text
    tokens = [stemmer.stem(word) for word in word_tokenize(text)]

    return tokens
```

```python
vectorizer = CountVectorizer(lowercase=True,
                             min_df=2,
                             max_df=0.9,
                             ngram_range=(1, 2),
                             max_features=5000,
                             tokenizer=tokenize,
                             stop_words=english_stopwords,
                             token_pattern=None)
```

```python
vectorizer.fit(df.text)
inputs = vectorizer.transform(df.text)
```

### Завдання 4. Побудова класифікатора

- Розділіть індекси даних на навчальний та тестовий набори в обраному співвівдношенні. Використовуючи отримані індекси сфомуйте набори для тренування класифікатора `X_train_bow, X_test_bow, y_train, y_test`.
- Навчіть класифікатор (наприклад, Logistic Regression, Decision Tree або один з алгоритмів бустингу) на даних, векторизованих методом bag-of-words. Спробуйте кілька моделей і оберіть найбільш точну :)
- Виведіть інформацію, яка дає можливість оцінити якість класифікації.
- Оцініть якість фінальної класифікації: вона хороша чи не дуже?



```python
# Compute F1 weighted, build confusion matrix
def compute_metrics(model, inputs, targets, name='', threshold=0.5):
  preds = model.predict(inputs)
  class_labels = model.classes_

  # Compute F1
  f1 = f1_score(targets, preds, average='weighted')
  print("F1 score: {:.2f}".format(f1))

  # Compute accuracy score
  accuracy = accuracy_score(targets, preds, normalize=True)
  print("Accuracy score: {:.2f}".format(accuracy))

  # Build confusion matrix
  cm_abs = confusion_matrix(targets, preds)
  cm_percentage = confusion_matrix(targets, preds, normalize='true')

  plt.figure()
  ax = sns.heatmap(cm_percentage, annot=True,
                   xticklabels=class_labels, yticklabels=class_labels)

  threshold = cm_abs.max() / 2
  for i in range(cm_abs.shape[0]):
    for j in range(cm_abs.shape[1]):
      ax.text(j + 0.5, i + 0.6, f'\n({cm_abs[i, j]})',
              ha='center', va='center', fontsize=10,
              color="white" if cm_abs[i, j] < threshold else "black")

  plt.xlabel('Prediction')
  plt.ylabel('Target')
  plt.title('{} Confusion Matrix'.format(name))
  plt.show()
```

```python
# Split indices into train and validation sets
train_indices, val_indices = train_test_split(df.index, stratify=df.sentiment,
                                              test_size=0.3, random_state=42)
```

```python
# Split the data into train and validation sets
train_inputs = inputs[train_indices]
train_targets = df.loc[train_indices]['sentiment']
val_inputs = inputs[val_indices]
val_targets = df.loc[val_indices]['sentiment']
```

**Logistic Regression**

```python
lr = LogisticRegression(solver='sag', max_iter=3000)
lr.fit(train_inputs, train_targets)
```

```python
compute_metrics(lr, train_inputs, train_targets, "Train")
compute_metrics(lr, val_inputs, val_targets, "Validation")
```

**One-vs-Rest with Logistic Regression**

```python
# One-vs-Rest multiclass classification
lr2 = LogisticRegression(solver='sag', max_iter=4000)
ovr_model = OneVsRestClassifier(lr2)
ovr_model.fit(train_inputs, train_targets)
```

```python
compute_metrics(ovr_model, train_inputs, train_targets, "Train")
compute_metrics(ovr_model, val_inputs, val_targets, "Validation")
```

**Decision Tree**

```python
params_dt = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best'],
    'max_depth': np.arange(5, 50, 5),
    'max_leaf_nodes': np.arange(10, 100, 10),
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 5, 10, 15],
    'max_features': [None, 'sqrt'],
}

cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
```

```python
random_search_tree = RandomizedSearchCV(
    DecisionTreeClassifier(random_state=42),
    params_dt,
    n_iter=100,
    cv=cv_strategy,
    scoring="f1_weighted",
    random_state=42,
    refit=True
)
```

```python
random_search_tree.fit(train_inputs, train_targets)
```

```python
dt = random_search_tree.best_estimator_
dt.fit(train_inputs, train_targets)
```

```python
compute_metrics(dt, train_inputs, train_targets, "Train")
compute_metrics(dt, val_inputs, val_targets, "Validation")
```

**XGB**

```python
# Encode target variable for XGB Classifier
label_encoder = LabelEncoder()
train_targets_encoded = label_encoder.fit_transform(train_targets)
val_targets_encoded = label_encoder.transform(val_targets)
```

```python
xgb = XGBClassifier(n_estimators=300,
    learning_rate=0.1,
    max_depth=12,
    min_child_weight=3,
    gamma=0.3,
    subsample=0.7,
    colsample_bytree=0.7,
    reg_alpha=1,
    reg_lambda=10)

xgb.fit(train_inputs, train_targets_encoded)
```

```python
compute_metrics(xgb, train_inputs, train_targets_encoded, "Train")
compute_metrics(xgb, val_inputs, val_targets_encoded, "Validation")
```

В цілому класифікатор є непоганим, адже правильно передбачає класи на валідаційному сеті в 69% випадків (тоді як якби ми завжди передбачали мажоритарний клас, точність класифікатора була би 40%). Кращої точності мені не вдалось досягнути, в експериментах з іншими параметрами моделі могли сильно перенавчитись, але не підвищували точність на валідаційному сеті.

### Завдання 5. Аналіз впливовості слів в отриманого класифікатора

- Для обраної вами моделі проведіть аналіз важливості слів (ознак): які слова (токени) найбільше впливають для визначення сентименту? Чи це логічно на ваш погляд, що саме ці символи впливають найбільше/найменще?


```python
def get_feature_importance_df(vectorizer, model):
  # Get feature names from the vectorizer
  feature_names = vectorizer.get_feature_names_out()

  # Get the feature importance from the model
  importances = model.feature_importances_

  # Create a DataFrame to display the feature importance with corresponding feature names
  feature_importance_df = pd.DataFrame({
      'feature': feature_names,
      'importance': importances
  })

  # Sort by importance (descending)
  feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

  return feature_importance_df
```

```python
# Display top 100 important features
feature_importance_bow = get_feature_importance_df(vectorizer, xgb)
feature_importance_bow.head(100)
```

Вплив цих слів виглядає доволі логічним, адже переважно це емоційно забарвлені слова, які впливають на сентимент тексту.
