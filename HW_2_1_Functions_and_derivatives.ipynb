### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function_and_derivative_plot(x_values, x, f, df):
  # convert the symbolic expressions to numerical functions using numpy
  f_num = sp.lambdify(x, f, 'numpy')
  df_num = sp.lambdify(x, df, 'numpy')

  # find values f(x)
  f_values = f_num(x_values)

  #find values df(x)
  is_constant = np.isscalar(df_num(x_values))
  if is_constant:
    # if the derivative is constant, create an array of constant values
    df_values = np.full_like(x_values, df_num(x_values))
  else:
    # otherwise evaluate the derivative at each x value
    df_values = df_num(x_values)

  # create a plot
  plt.figure(figsize=(10, 5))
  plt.plot(x_values, f_values, label="f(x) = " + str(f), color='blue')
  plt.plot(x_values, df_values, label="df(x) = " + str(df), color='green')

  # add legend, title, labels
  plt.title("Function and its derivative")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = 10 * x - 5
df = sp.diff(f, x)

print("Derivative of a linear function 10 * x - 5:", df)

# generate a range of values
x_values = np.arange(0, 11, 1)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```

–í–∏—Å–Ω–æ–≤–æ–∫:

–§—É–Ω–∫—Ü—ñ—è –ª—ñ–Ω—ñ–π–Ω–∞, –æ—Ç–∂–µ –≤–æ–Ω–∞ –∑—Ä–æ—Å—Ç–∞—î –ª—ñ–Ω—ñ–π–Ω–æ —Ä–∞–∑–æ–º –∑—ñ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è–º –∑–º—ñ–Ω–Ω–æ—ó: –∑ –∫–æ–∂–Ω–æ—é –¥–æ–¥–∞—Ç–∫–æ–≤–æ—é –≥–æ–¥–∏–Ω–æ—é –Ω–∞–≤—á–∞–Ω–Ω—è –æ—Ü—ñ–Ω–∫–∞ –∑–±—ñ–ª—å—à—É–≤–∞—Ç–∏–º–µ—Ç—å—Å—è –Ω–∞ 10 –±–∞–ª—ñ–≤.

–ü–æ—Ö—ñ–¥–Ω–∞ —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó —î –∫–æ–Ω—Å—Ç–∞–Ω—Ç–æ—é, –æ—Ç–∂–µ —Ñ—É–Ω–∫—Ü—ñ—è –º–∞—î –ø–æ—Å—Ç—ñ–π–Ω—É —à–≤–∏–¥–∫—ñ—Å—Ç—å –∑–º—ñ–Ω–∏.

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



**ùëì(ùë•)=sin(ùë•)**

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = sp.sin(x)
df = sp.diff(f, x)

print("Derivative of a function sin(x): ", df)

# generate a range of values
x_values = np.linspace(-10, 10, 100)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```

**ùëì(ùë•)=ln(ùë•)**

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = sp.log(x)
df = sp.diff(f, x)

print("Derivative of a function ln(x): ", df)

# generate a range of values
x_values = np.linspace(-10, 10, 100)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```


3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = x**2 - 5 * x + 1
df = sp.diff(f, x)

print("Derivative of a function x**2 - 5 * x + 1: ", df)

# generate a range of values
x_values = np.arange(0, 100, 1)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```
