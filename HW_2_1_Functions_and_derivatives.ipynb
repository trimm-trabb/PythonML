1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function_and_derivative_plot(x_values, x, f, df):
  # convert the symbolic expressions to numerical functions using numpy
  f_num = sp.lambdify(x, f, 'numpy')
  df_num = sp.lambdify(x, df, 'numpy')

  # find values f(x)
  f_values = f_num(x_values)

  #find values df(x)
  is_constant = np.isscalar(df_num(x_values))
  if is_constant:
    # if the derivative is constant, create an array of constant values
    df_values = np.full_like(x_values, df_num(x_values))
  else:
    # otherwise evaluate the derivative at each x value
    df_values = df_num(x_values)

  # create a plot
  plt.figure(figsize=(10, 5))
  plt.plot(x_values, f_values, label="f(x) = " + str(f), color='blue')
  plt.plot(x_values, df_values, label="df(x) = " + str(df), color='green')

  # add legend, title, labels
  plt.title("Function and its derivative")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = 10 * x - 5
df = sp.diff(f, x)

print("Derivative of a linear function 10 * x - 5:", df)

# generate a range of values
x_values = np.arange(0, 11, 1)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```

Висновок:

Функція лінійна, отже вона зростає лінійно разом зі зростанням змінної: з кожною додатковою годиною навчання оцінка збільшуватиметься на 10 балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



**𝑓(𝑥)=sin(𝑥)**

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = sp.sin(x)
df = sp.diff(f, x)

print("Derivative of a function sin(x): ", df)

# generate a range of values
x_values = np.linspace(-10, 10, 100)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```

**𝑓(𝑥)=ln(𝑥)**

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = sp.log(x)
df = sp.diff(f, x)

print("Derivative of a function ln(x): ", df)

# generate a range of values
x_values = np.linspace(-10, 10, 100)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# define a variable
x = sp.symbols('x')

# define function and its derivative
f = x**2 - 5 * x + 1
df = sp.diff(f, x)

print("Derivative of a function x**2 - 5 * x + 1: ", df)

# generate a range of values
x_values = np.arange(0, 100, 1)

# draw a plot
function_and_derivative_plot (x_values, x, f, df)
```
